02) Arrume-o: troque o nome da coluna ‘referencia’ por data e a coloque
como índice do Data Frame. Exclua a coluna ‘periodo’. Os valores de
‘pobreza’ e ‘extrema_pobreza’ estão invertidos, troque-os.

```python
dados = dados.rename(columns={"referencia": "data"})
dados = dados.set_index("data")
dados = dados.drop(["periodo"], axis = 1)
dados = dados.rename(columns={"pobreza": "temp", "extrema_pobreza": "pobreza"})
dados = dados.rename(columns={"temp": "extrema_pobreza"})
```
